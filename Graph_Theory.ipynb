```python
import pandas as pd
```

```python
df = pd.read_csv("graphdata.csv")
```

```python
df.head()
```

```python
import networkx as nx
df = nx.from_pandas_edgelist(df, source="Origin", target="Dest", edge_attr=True)
```

```python
df.nodes()
```

```python
df.edges()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(12,8))
nx.draw_networkx(df, with_labels=True)
```
